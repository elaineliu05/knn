````python
from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test

import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5):
    print(y_test[i],'\n') # Print the label
    print(X_test[i],'\n') # Print the matrix of values
    plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
    plt.show()
````

````python
#2.
print(X_train.shape)
print(X_test.shape)
print(X_train[0].shape)
print(X_test[0].shape)
print(y_train.shape)
print(y_test.shape)
````

````python
#3.
import pandas as pd

X_train_reshaped = X_train.reshape((60000,784))
X_test_reshaped = X_test.reshape((10000,784))
````

````python
#4.
from sklearn.neighbors import KNeighborsClassifier

accuracy = []
for k in range(11):
    knn = KNeighborsClassifier(n_neighbors=k+1) # Create classifier instance
    predictor = knn.fit(X_train_reshaped,y_train) # Fit model
    accuracy.append( predictor.score(X_test_reshaped,y_test) ) # Store accuracy; bug in sklearn requires .values

accuracy_max = np.max(accuracy) # highest recorded accuracy
max_index = np.where(accuracy==accuracy_max)[0]
k_star = max_index+1 # Find the optimal value of k
print('optimal k: ', k_star)

plt.plot(np.arange(1,12),accuracy) # Plot accuracy by k
plt.xlabel("k")
plt.ylabel('Accuracy')
plt.show()
````

````python
#5.
knn = KNeighborsClassifier(n_neighbors=3) # Create classifier instance
predictor = knn.fit(X_train_reshaped,y_train) # Fit model
y_hat = predictor.predict(X_test_reshaped)
print('accuracy: ', accuracy[2])

pd.crosstab(y_test, y_hat)

print('when it chooses 3 nearest neighbors, the accuracy is 97%, which is very high. The model tends to confuse 7s for 1s and 4s for 9s, which makes sense.')
````

````python
#6.
print('The mistakes the model is most likely to make is confusing a 7 for a 1, a 4 for a 9, and a 8 for a 3. These are all '+
      'numbers that have similar shapes, so it makes sense. No crazy interesting patterns other than this. ')
````
